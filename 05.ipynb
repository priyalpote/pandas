#Grouping and Aggregarion

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("marketing_campaign.csv", sep='\t')
df.info()
```

#Average amount bought by each group

```python
amt_bought=['MntWines'     ,  'MntFruits'          ,  'MntMeatProducts'   ,  'MntFishProducts'  ,  'MntSweetProducts','MntGoldProds']
```

```python
df.head()
```

```python
df.columns
```

```python
df.groupby(['Marital_Status']).mean()[amt_bought]
```

#Multiple columns

```python
df.groupby(['Education','Marital_Status'])['Income'].agg(['median','mean'])
```

#Custom agreegation

```python
def getTop5(df, n=5, col='NumWebPurchases'):
  return df.sort_values(by=col)[-1:-6:-1]

getTop5(df)
```

```python
visits = ['NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']
df.groupby(['Education']).apply(getTop5)[visits]
```

#Stock returns

```python
df= pd.read_csv('stock_data.csv', index_col='Unnamed: 0', parse_dates=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.value_counts()
```

```python
ret=df.pct_change().dropna()
ret
```

```python
getYr= lambda x: x.year
byYrStocks= ret.groupby(getYr).sum()*100
```
