#Filtering Data

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("marketing_campaign.csv", sep='\t')
df.head()
```

```python
df.info()
```

```python
df['Marital_Status'].value_counts()
```

```python
df[df['Marital_Status']=='Married']
```

#Not Married , not togather and born after 1990

```python
df.loc[~(df['Marital_Status'].isin(['Married','Together'])) & (df['Year_Birth']>1990)]
```

#bitwise operator to filter data

```python
df.loc[(df['NumWebPurchases'] >=5) | (df['NumDealsPurchases'] >=5) , 'Active online buyers' ] = 'Active'
df.loc[(df['NumWebPurchases'] <5) & (df['NumDealsPurchases'] <5) , 'Active online buyers' ] = 'Not Active'
```
